```python
hippocampus_genes = pd.read_csv(hippocampus_file, sep="\t")
```

```python
# for gene_id in set(hippocampus_genes["gene_id"]):
#     print(gene_id)
```

```python
# unique_hc_genes contains all the Ensembl Gene Ids
# for genes that affect the hippocampus

unique_hc_genes = list(map(lambda x: x.split(".")[0], set(hippocampus_genes["gene_id"])))
len(unique_hc_genes)
```

```python
gene_info = m.getgenes(unique_hc_genes)
```

```python
len(gene_info)
```

```python
hippocampus_entrez = []
for info in gene_info:
    if "entrezgene" not in info.keys():
        continue
    hippocampus_entrez.append(info["entrezgene"])
```

```python
open("hippocampus_genes.txt", "w").write("\n".join(map(str, hippocampus_entrez)))
```
