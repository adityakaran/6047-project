```python
import interactome as I
ixome = I.Interactome(I.PATH + I.INTERACTOME_FILE)
```

```python
hippocampus_genes = list(map(int, open("./tissue-genes-and-snps/hippocampus_entrez.txt").read().split("\n")))
```

```python
hippocampus_genes_in_ixome = ixome.subgraph(hippocampus_genes, filter=True)
```

```python
len(hippocampus_genes), len(hippocampus_genes_in_ixome)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(12, 12))
plt.title("Hippocampus Genes")
nx.draw_networkx(hippocampus_genes_in_ixome, with_labels=False, node_size=3)
```

```python
from disease import Disease
```

```python
schizo = Disease("schizophrenia", ixome, hippocampus_genes_in_ixome.nodes)
```

```python
nx.draw_networkx(schizo.subgraph, with_labels=False, node_size=3)
```

```python
pos = nx.spring_layout(hippocampus_genes_in_ixome)
plt.figure(figsize=(12, 12))
nx.draw_networkx_nodes(hippocampus_genes_in_ixome, pos, node_color="r", node_size=3, alpha=0.3)
nx.draw_networkx_nodes(hippocampus_genes_in_ixome, pos, nodelist=list(schizo.subgraph.nodes), node_color="b", alpha=1.0, node_size=12)
nx.draw_networkx_edges(hippocampus_genes_in_ixome, pos, width=1.0, alpha=0.3)
```

```python
diabetes = Disease("diabetes", ixome, hippocampus_genes_in_ixome.nodes)
plt.figure(figsize=(12, 12))
nx.draw_networkx_nodes(hippocampus_genes_in_ixome, pos, node_color="r", node_size=3, alpha=0.3)
nx.draw_networkx_nodes(hippocampus_genes_in_ixome, pos, nodelist=list(diabetes.subgraph.nodes), node_color="b", alpha=1.0, node_size=12)
nx.draw_networkx_edges(hippocampus_genes_in_ixome, pos, width=1.0, alpha=0.3)
```

```python
autism = Disease("autism", ixome, hippocampus_genes_in_ixome.nodes)
plt.figure(figsize=(12, 12))
nx.draw_networkx_nodes(hippocampus_genes_in_ixome, pos, node_color="r", node_size=3, alpha=0.3)
nx.draw_networkx_nodes(hippocampus_genes_in_ixome, pos, nodelist=list(autism.subgraph.nodes), node_color="b", alpha=1.0, node_size=12)
nx.draw_networkx_edges(hippocampus_genes_in_ixome, pos, width=1.0, alpha=0.3)
```
